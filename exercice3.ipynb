```python
def point_fixe(f,x0,ep,nmax):
    a=x0
    for i in range(nmax):
        x=a
        a=f(x)
        erreur=abs(a-x)
        print(a)
        if erreur<ep:
            return a,i
    return "la fonction est diverge"
```

```python
import numpy as np
x, y, z = symbols('x y z')
init_printing(use_unicode=True)
g1=lambda x: 1/2*sqrt(10-x**3)
g2=lambda x: sqrt(10/(x+4))
g3=lambda x: x-x**3-4*x**2+10
```

```python
point_fixe(g2,1.5,0.001,40)
```

```python
point_fixe(g1,1.5,0.001,40)
```
